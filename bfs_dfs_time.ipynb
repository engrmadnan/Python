```python
from collections import deque
import time
graph = {
    'Karachi': ['Clifton', 'Saddar', 'Gulshan', 'Korangi'],
    'Clifton': ['DHA', 'Sea View'],
    'Saddar': ['Tower', 'Empress Market'],
    'Gulshan': ['Gulshan-e-Iqbal', 'Nipa'],
    'Korangi': ['Landhi', 'Shah Faisal'],
    'DHA': ['Khyaban-e-Ittehad', 'Beach Park'],
    'Gulshan-e-Iqbal': ['University Road', 'National Stadium'],
    'Landhi': ['Industrial Area', 'Cattle Colony'],
    'Shah Faisal': ['Airport', 'Malir'],
    'Tower': ['Quaid Mausoleum'],
    'Empress Market': ['Burns Road'],
    'National Stadium': ['Hassan Square']
}

def bfs(graph, start):
    visited_node=set()
    initial_queue=deque([(start,[start])])
    start_time=time.process_time()
    path =[]
    while initial_queue:
        current_node,current_path=initial_queue.popleft()
        for nearest in graph.get(current_node,[]):
            if nearest not in visited_node:
                visited_node.add(nearest)
                new_path = current_path+[nearest]
                initial_queue.append((nearest,new_path))
                path.append(new_path)
    end_time=time.process_time()
    total_time_bfs=end_time-start_time
    return new_path, total_time_bfs
#	return new_path             for shortest path last location only

def dfs(graph, start):
    visited_node=set()
    initial_stack=([(start,[start])])
    start_time=time.process_time()
   # path =[]
    while initial_stack:
        current_node,current_path=initial_stack.pop()
        if current_node not in visited_node:
            visited_node.add(current_node)
            for nearest in graph.get(current_node,[]):



                if nearest not in visited_node:

                    new_path = current_path+[nearest]

                    initial_stack.append((nearest,new_path))
                  #  path.append(new_path)
    end_time=time.process_time()
    total_time_dfs=end_time-start_time
    return current_path, total_time_dfs
#	return new_path             for shortest path last location only

while True:

    print ("\nAlgorithms")
    print ("1- For BFS ")
    print ("2- For DFS")
    print ("3- For Optimal Time")
    print ("4- Exit")


    choice = input("Enter the choice between (1-4)")

    if choice=='1':
        all_path,total_time_bfs=bfs(graph,'Karachi')
        print(total_time_bfs)
        for path in all_path:
            print(path)

    elif choice=='2':
        final_output, total_time_dfs=dfs(graph,'Karachi')
        print(final_output)
        print(total_time_dfs)

    elif choice=='3':
        all_path,total_time_bfs=bfs(graph,'Karachi')
        print("BFS Time = " ,total_time_bfs)

        final_output, total_time_dfs=dfs(graph,'Karachi')
        print("DFS Time = ",total_time_dfs)

        if (total_time_bfs>total_time_dfs):
            print("Optimal Solution is DFS ")
        else:

            print("Optimal Solution is BFS")

    elif choice=='4':
        print("Thanks for using ")
        break

    else:
        print("Invalid Choice! Please enter a valid number between (1-4) again: ")














```
